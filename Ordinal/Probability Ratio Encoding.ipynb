# Probability Ratio Encoding

```python
import pandas as pd

data = pd.read_csv('titanic.csv',usecols=['Cabin','Survived'])
data.head()
```

```python
data['Cabin'].fillna('Missing',inplace=True)
data.head()
```

```python
data['Cabin'] = data['Cabin'].astype(str).str[0]
data.head()
```

```python
data['Cabin'].unique()
```

```python
prob_data = data.groupby(['Cabin'])['Survived'].mean()
prob_data
```

```python
prob_data=pd.DataFrame(prob_data)
prob_data
```

```python
prob_data['Died'] = 1-prob_data['Survived']
prob_data.head()
```

```python
prob_data['prob_ratio'] = prob_data['Survived'] / prob_data['Died']
prob_data.head()
```

```python
prob_ratio = prob_data['prob_ratio'].to_dict()
```

```python
data['Cabin_Encoded'] = data['Cabin'].map(prob_ratio)
data.head()
```
